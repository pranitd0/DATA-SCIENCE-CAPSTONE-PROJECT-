```python

# Task 1: Mark all launch sites on a map
# Get unique launch sites and their coordinates
launch_sites = spacex_df.groupby(['Launch Site', 'Lat', 'Long'], as_index=False).first()

# Define the starting point of the map
center_lat = launch_sites.loc[0, 'Lat']
center_lon = launch_sites.loc[0, 'Long']

# Create a base Folium map object
site_map = folium.Map(location=[center_lat, center_lon], zoom_start=4)

# Iterate through the unique launch sites and add markers
for index, site in launch_sites.iterrows():
    location_coords = [site['Lat'], site['Long']]
    
    # Create a Marker for the site
    marker = folium.Marker(
        location=location_coords,
        icon=folium.Icon(color='red', icon='info-sign'),
        popup=site['Launch Site']
    )
    
    # Add the marker to the map
    marker.add_to(site_map)

# Save the map to an HTML file for viewing
site_map.save("spacex_launch_sites_task1.html")
```

```python

# Task 2: Mark the success/failed launches for each site on the map
# Create a base map (reusing logic from Task 1 to find center)
launch_sites = spacex_df.groupby(['Launch Site', 'Lat', 'Long'], as_index=False).first()
center_lat = launch_sites.loc[0, 'Lat']
center_lon = launch_sites.loc[0, 'Long']
site_map_task2 = folium.Map(location=[center_lat, center_lon], zoom_start=5)

# Create a MarkerCluster object
marker_cluster = MarkerCluster()
site_map_task2.add_child(marker_cluster)

# Iterate through the DataFrame and add a marker to the MarkerCluster for each launch
for index, record in spacex_df.iterrows():
    # Determine the marker color based on the 'class' column (0=Failure, 1=Success)
    if record['class'] == 1:
        color = 'green'
    else:
        color = 'red'
        
    # Create a Marker object with the determined color and add it to the MarkerCluster
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],
        icon=folium.Icon(color='white', icon_color=color),
        popup=f"Launch Site: {record['Launch Site']}<br>Outcome: {'Success' if record['class'] == 1 else 'Failure'}"
    )
    marker_cluster.add_child(marker)

# Save the map to an HTML file
site_map_task2.save("spacex_launch_sites_task2.html")
```

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

```python
import folium
from folium.plugins import MarkerCluster

# Assuming 'site_map' is already defined from the previous steps
# Assuming 'spacex_df' is the detailed DataFrame with 'Launch Site', 'Lat', 'Long', and 'class' columns

# Add the MarkerCluster object to the map
site_map.add_child(marker_cluster)

# Iterate through the spacex_df DataFrame
for index, record in spacex_df.iterrows():
    # Define the marker coordinate
    coordinate = [record['Lat'], record['Long']]
    
    # Determine the color based on the 'class' value
    if record['class'] == 1:
        color = 'green'
        outcome = 'Successful Launch'
    else:
        color = 'red'
        outcome = 'Failed Launch'
        
    # Create a Marker with the appropriate color and a popup showing the outcome
    marker = folium.Marker(
        location=coordinate,
        icon=folium.Icon(color=color, icon='info-sign'), # Use a descriptive icon
        popup=f"{record['Launch Site']}: {outcome}" # Add launch details to the popup
    )
    
    # Add the marker to the MarkerCluster
    marker_cluster.add_child(marker)

# Display the map with the clusters
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import sin, cos, sqrt, atan2, radians

# Earth's radius in kilometers
R = 6371

def calculate_distance(lat1, lon1, lat2, lon2):
    # convert decimal degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    # haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance  


    # Launch site coordinates
launch_site_lat = 28.563197
launch_site_lon = -80.576820

# Nearest Highway coordinates (Example point)
highway_lat = 28.56339
highway_lon = -80.56784

distance_highway = calculate_distance(launch_site_lat, launch_site_lon, highway_lat, highway_lon)

print(f"The distance from CCAFS SLC-40 to the nearest highway is: {distance_highway:.2f} km")


# Coordinate of the proximity point (Highway)
highway_coordinate = [highway_lat, highway_lon]

# Add a marker for the highway point
folium.Marker(
    highway_coordinate,
    icon=folium.Icon(color='blue', icon='road', prefix='fa'),
    popup=f"Nearest Highway ({distance_highway:.2f} km)"
).add_to(site_map)

# Draw a line between the launch site and the highway
folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], highway_coordinate], 
    weight=2, 
    color='blue'
).add_to(site_map)

# Display the map (assuming site_map is the Folium map object from Task 2)
site_map
```
